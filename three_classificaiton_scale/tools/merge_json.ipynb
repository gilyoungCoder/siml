```python
import os
import json

def merge_json_files(directory, output_file):
    merged_data = []
    
    # 디렉토리 내 모든 파일 확인
    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            filepath = os.path.join(directory, filename)
            
            with open(filepath, 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
                merged_data.append(data)
    
    # 합쳐진 데이터를 하나의 JSON 파일로 저장
    with open(output_file, 'w', encoding='utf-8') as output_json_file:
        json.dump(merged_data, output_json_file, ensure_ascii=False, indent=4)

# 사용 예시
directory = '/mnt/home2/etri/data/train/annotations'  # JSON 파일들이 있는 디렉토리 경로
output_file = '/mnt/home2/etri/data/train/annotations.json'  # 합쳐진 JSON 파일의 경로 및 이름

merge_json_files(directory, output_file)
```

```python
import os
import json

def convert_bbox(yolo_bbox):
    center_x, center_y, width, height = yolo_bbox
    top_left_x = center_x - width / 2
    top_left_y = center_y - height / 2
    bottom_right_x = center_x + width / 2
    bottom_right_y = center_y + height / 2
    return [top_left_x, top_left_y, bottom_right_x, bottom_right_y]

def merge_and_convert_json_files(directory, output_file):
    merged_data = {
        "images": [],
        "annotations": [],
        "categories": []
    }
    annotation_id = 1
    image_id_set = set()
    
    # 디렉토리 내 모든 파일 확인
    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            filepath = os.path.join(directory, filename)
            
            with open(filepath, 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
                
                # 이미지 정보 추가
                for image in data["images"]:
                    if image["id"] not in image_id_set:
                        merged_data["images"].append(image)
                        image_id_set.add(image["id"])
                
                # 어노테이션 변환 및 추가
                for annotation in data["annotations"]:
                    converted_bbox = convert_bbox(annotation["bbox"])
                    annotation["bbox"] = converted_bbox
                    annotation["id"] = annotation_id
                    merged_data["annotations"].append(annotation)
                    annotation_id += 1
                
                # 카테고리 정보 추가 (중복 확인)
                for category in data["categories"]:
                    if category not in merged_data["categories"]:
                        merged_data["categories"].append(category)
    
    # 합쳐진 데이터를 하나의 JSON 파일로 저장
    with open(output_file, 'w', encoding='utf-8') as output_json_file:
        json.dump(merged_data, output_json_file, ensure_ascii=False, indent=4)

# 사용 예시
directory = '/mnt/home2/etri/data/train/annotations'  # JSON 파일들이 있는 디렉토리 경로
output_file = '/mnt/home2/etri/data/train/annotations.json'  # 합쳐진 JSON 파일의 경로 및 이름

merge_and_convert_json_files(directory, output_file)
```
