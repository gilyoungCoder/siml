```python
# # Remove half of the log
# sorted_unseen_log = sorted_unseen_log[:len_of_selected]

# sorted_unseen_log = dict(sorted_unseen_log)

# with open("unseen_selected_cycle_2.yaml", "w") as f:
#     yaml.dump(sorted_unseen_log, f)

random_index_select = random.sample(range(len(unseen_log)), len_of_selected)
new_selected_dict = {}

count = 0
for i in unseen_log:
    if count in random_index_select:
        new_selected_dict[i] = unseen_log[i]
    count += 1

len(new_selected_dict)
```
